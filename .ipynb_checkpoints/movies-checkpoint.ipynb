```python
extrema = lambda x: (np.min(x), np.max(x))
```

```python
%matplotlib notebook
ds = xr.open_dataset("/Users/anthonymeza/Downloads/bsose_i105_2008to2012_monthly_SSH.nc")
ETAN = ds.ETAN
vmin, vmax = extrema(ETAN)
ETAN = ETAN.where(ETAN != 0)
fig, ax = plt.subplots()
ETAN.isel(time = 0).plot(ax = ax)

fig
```
